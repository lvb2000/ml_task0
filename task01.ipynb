```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error
```

## include data set

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
train
```

## convert datafram to numpy 

```python
y_df=train.y
y= y_df.to_numpy()
X=train.to_numpy()[:,2:-1]
```

## closed form solution

```python
X_T=np.transpose(X)
w=np.linalg.inv(X_T@X)@X_T@y
w
```

## eval test set

```python
X_test=test.to_numpy()[:,1:-1]
y_final=X_test@w
y_final
```

```python
Id=(test.Id).to_numpy()
arr=np.array([Id,y_final]).T
submission = pd.DataFrame(data=arr, columns=['Id','y'])
submission.to_csv("submission_taks0")
```
